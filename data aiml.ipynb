```python
# Step 1: Import Libraries
# ==========================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler, PolynomialFeatures
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
```

```python
df=pd.read_csv("data.csv")
print("Shape:",df.shape)
print("Columns:", df.columns.tolist())
print("\nMissing values:\n", df.isnull().sum())
print("\nDuplicate rows:\n", df.duplicated().sum())
```

```python
#step 2: Drop Irrelevant columns
df = df.drop(['date', 'street', 'city', 'statezip', 'country'], axis=1)
```

```python
#Step 3: Handle Missing value
#Median use Middile value to fill missing value
#Object means we use mode(Repeated values fill)
for col in df.select_dtypes(include=['int64','float64']).columns:
    df[col].fillna(df[col].median(),inplace=True)
for col in df.select_dtypes(include=['object']).columns:
    df[col].fillna(df[col].mode()[0],inplace=True)
```

```python
#Step 4: Encode Categorical features (Label Encoding: String(objects) convert to integers)
label_encoders ={}
for col in df.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le
```

```python
#Step 5: Feature Selection
target_column = "price"
X=df.drop(target_column,axis=1)
y = df[target_column]
```

```python
#Step 6: Feature Scaling
scalar = StandardScaler()
X_Scaled = scalar.fit_transform(X)
```

```python
#Step 7: Train-Test_Split
x_train,x_test, y_train,y_test = train_test_split(X_Scaled,y,test_size=0.2,random_state=42)
```

```python
#Step: 8
def evaluate_model(y_true,y_pred):
    mae = mean_absolute_error(y_true,y_pred)
    mse = mean_squared_error(y_true,y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_true,y_pred)
    print("MAE:",mae)
    print("MSE:",mse)
    print("RMSE:",rmse)
    print("R² Score:",r2)
```

```python
#linear regression 
print("\n===== Linear Regression =====")
lr= LinearRegression()
lr.fit(x_train, y_train)
y_pred = lr.predict(x_test)
evaluate_model(y_test, y_pred)
```

```python
#multiple linear regression
print("\n===== Multiple Linear Regression =====")
mlr = LinearRegression()
mlr.fit(x_train, y_train) 
y_pred =mlr.predict(x_test)
evaluate_model(y_test, y_pred)
```

```python
#polynomial linear regression
print("\n===== Polynomial Regression =====")
poly = PolynomialFeatures(degree=2)
x_train_poly = poly.fit_transform(x_train)
x_test_poly = poly.transform(x_test)

poly_model = LinearRegression()
poly_model.fit(x_train_poly, y_train)
y_pred = poly_model.predict(x_test_poly)
evaluate_model(y_test, y_pred)
```

```python
#ridge regression
print("\n===== Ridge Regression =====")
ridge = Ridge(alpha=1.0)
ridge.fit(x_train, y_train)
y_pred = ridge.predict(x_test)
evaluate_model(y_test, y_pred)
```

```python
#5.Lasso Regression
print("\n===== Lasso Regression =====")
laaso = Lasso(alpha=0.1)
laaso.fit(x_train, y_train)
y_pred = laaso.predict(x_test)
evaluate_model(y_test, y_pred)
```

```python
#6. Elastic Net Regression
print("\n===== Elastic Net Regression =====")
elastic_net = ElasticNet(alpha=0.1, l1_ratio=0.5)
elastic_net.fit(x_train, y_train)
y_pred = elastic_net.predict(x_test)
evaluate_model(y_test, y_pred)
```

```python
#7. Decision Tree Regression
print("\n===== Decision Tree Regression =====")
dt = DecisionTreeRegressor(max_depth=5)
dt.fit(x_train, y_train)
y_pred = dt.predict(x_test)
evaluate_model(y_test, y_pred)
```

```python
#random forest regression
print("\n===== Random Forest Regressor =====")
rf = RandomForestRegressor(n_estimators=100, max_depth=5, random_state=42)
rf.fit(x_train, y_train)    
y_pred = rf.predict(x_test)
evaluate_model(y_test, y_pred)
```

```python
#support vector regression
print("\n===== Support Vector Regression =====")
svr = SVR(kernel='rbf', C=100, gamma='scale')
svr.fit(x_train, y_train)
y_pred = svr.predict(x_test)
evaluate_model(y_test, y_pred)
```
