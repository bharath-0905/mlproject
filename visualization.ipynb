```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("CAR DETAILS FROM CAR DEKHO.csv")
df
```

```python
df.info()
```

```python
#count plot

plt.Figure(figsize=(8,6)) #Length and breadth
sns.countplot(x='seller_type',data=df,palette='viridis')
plt.title('selling_type count plot')
plt.xlabel('seller type')
plt.ylabel('count')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(df['year'].dropna(),bins=30,kde=True,color='green')
plt.title('Hisplot for year')
plt.xlabel('year')
plt.ylabel('count')
plt.show()
```

```python
#piechart

cars = df['owner'].value_counts()

plt.figure(figsize=(8,6))
plt.pie(cars, labels=['First Owner','Second Owner','Third Owner','Fourth and other','Test Drive Car'],
        autopct = '%1.1f%%',colors=['lightcoral','lightgreen','red','blue','yellow'],
        startangle=180,explode = (0.05,0.05,0.05,0.05,0.05))
plt.title('Something')
plt.show()
```

```python
#Box plot(To check outlayers(unrelevant data in our dataset) in our data)

plt.figure(figsize=(8,6))
sns.boxplot(x='seller_type',y='selling_price',data=df,palette='viridis')
plt.xlabel('class')
plt.ylabel('Fare')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='year',y='km_driven',data=df,hue='owner',style='seller_type',palette='rainbow')
plt.title('Scatter plot of year vs km_driven with owner and seller_type ')
plt.xlabel('year')
plt.ylabel('km_driven')
plt.legend(title='km_driven')
plt.show()
```

```python
#violin plot(Combination of box and destiny plot)

plt.figure(figsize=(10,8))
sns.violinplot(x='fuel',y='year',hue='seller_type',data=df,
               split=True,palette='Set2')
plt.title('seller_type Distribution by year and fuel')
plt.xlabel('fuel')
plt.ylabel('cars')
plt.legend(title = 'Cars')
plt.show()
```

```python
sns.kdeplot(data=df['year'].dropna(),color='skyblue',shade=True)
plt.title('year Distribution(kernal Density Estimate)')
plt.xlabel('year')
plt.ylabel('Density')
plt.show()

```

```python
#Face grid plot

g = sns.FacetGrid(df,col='seller_type',row='owner',margin_titles=True)
g.map(sns.histplot,'year',bins=20,color='teal')
g.fig.suptitle('car fuel',y=1.03)
plt.show()
```

```python
sns.pairplot(df[['year','selling_price','km_driven','transmission']],hue='transmission',palette='husl')
plt.suptitle('Pairplot of year, selling_price, km_driven and transmission',y=1.02)
plt.show()
```
